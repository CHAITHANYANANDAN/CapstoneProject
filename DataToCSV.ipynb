```python


def convert_mat_to_csv(mat_dir, csv_dir):
    
    # Create the CSV directory if it doesn't exist
    if not os.path.exists(csv_dir):
        os.makedirs(csv_dir)

    # Iterate through each MAT file in the directory
    for mat_file in os.listdir(mat_dir):
        if mat_file.endswith('.mat'):
            mat_path = os.path.join(mat_dir, mat_file)

            try:
                # Load the MAT file
                mat = sc.loadmat(mat_path)

                # Process each variable in the MAT file
                for var_name, data in mat.items():
                    if not var_name.startswith('__'):
                        # Convert the data to a DataFrame
                        df = pd.DataFrame(data)

                        # Construct the CSV file path
                        csv_file = f"{os.path.splitext(mat_file)[0]}_{var_name}.csv"
                        csv_path = os.path.join(csv_dir, csv_file)

                        # Save the DataFrame to a CSV file
                        df.to_csv(csv_path, index=False)
                        print(f"Saved {csv_path}")
            except Exception as e:
                print(f"Failed to process {mat_file}: {e}")
```

```python
import os
import pandas as pd

def find_shape_of_csv_recursive(csv_dir):
    """
    Find the shape of each CSV file in the specified directory and its subdirectories.

    Parameters:
    csv_dir (str): Directory containing the CSV files and subdirectories.

    Returns:
    dict: A dictionary with the CSV file paths as keys and their shapes (rows, columns) as values.
    """
    shapes = {}

    # Walk through the directory and its subdirectories
    for root, dirs, files in os.walk(csv_dir):
        for file in files:
            if file.endswith('.csv'):
                csv_path = os.path.join(root, file)
                
                try:
                    # Read the CSV file
                    df = pd.read_csv(csv_path)

                    # Get the shape of the DataFrame
                    shape = df.shape
                    
                    # Store the result in the dictionary
                    shapes[csv_path] = shape
                    print(f"{csv_path}: {shape[0]} rows, {shape[1]} columns")
                except Exception as e:
                    print(f"Failed to process {csv_path}: {e}")

    return shapes


```
